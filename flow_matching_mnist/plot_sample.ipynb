```python
import matplotlib.pyplot as plt
import torch
```

```python
from trainers.mnist import FlowMatching

# Load the trained model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
flow_matching_model = FlowMatching(device=device)

flow_matching_model.model.load_state_dict(torch.load('flow_matching_mnist.pth', map_location=device))

final_samples = flow_matching_model.sample(n_samples=64, n_steps=100)
    
    
samples = final_samples.cpu().numpy()
fig, axes = plt.subplots(8, 8, figsize=(12, 12))
for i, ax in enumerate(axes.flat):
    if i < len(samples):
        ax.imshow(samples[i, 0], cmap='gray')
        ax.axis('off')

plt.suptitle('Final Generated MNIST Samples')
plt.tight_layout()
plt.show()
```
